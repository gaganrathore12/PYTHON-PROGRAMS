```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
x= np.random.normal(0, 1, 1000)  # Mean = 0, Std Dev = 1, 1000 samples

# Create histogram
plt.hist(x, bins=20, color='skyblue', edgecolor='black', alpha=0.7, label='Data')

# Add labels, title, and gridlines
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Customized Histogram Example')
plt.grid(True)  # Adding gridlines
plt.legend()  # Adding legend

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
# Data
labels = ['Group 1', 'Group 2', 'Group 3', 'Group 4']
colors = ['blue', 'orange', 'green', 'red']
# Plot pie chart
plt.pie(x,sizes=10, labels=labels, autopct='%1.1f%%', startangle=140,colors=colors)
# Add title
plt.title('Pie Chart Example')

# Display the chart
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
w= 0.2
Categories = ['Category A', 'Category B', 'Category C', 'Category D']  
Boys =[20, 35, 30, 35]  
Girls =[25, 32, 34, 20]
Others =[15, 27, 29, 31]  

bar1= np.arange(len(Categories))
plt.bar(bar1,Boys,w,color="lightcoral", label="Boys")
bar2 = [i+w for i in bar1]
plt.bar(bar2,Girls,w,color = "tan", label="Girls")

bar3 = [i+w for i in bar2]
plt.bar(bar3,Others,w,color = "aquamarine",label="Others")
plt.xticks(bar1+w/2,Categories)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
w = 0.2
Categories = ['Category A', 'Category B', 'Category C', 'Category D']  
Boys =[20, 35, 30, 35]  
Girls =[25, 32, 34, 20]
Others =[15, 27, 29, 31]
y = [28, 30, 24, 36]  

bar1 = np.arange(len(Categories))
plt.bar(bar1,Boys,w,color="chocolate",edgecolor="black",label = "Boys")
bar2 = [i+w for i in bar1]
plt.bar(bar2,Girls,w,color="salmon",edgecolor="black",label = "Girls")
bar3  = [i+w for i in bar2]
plt.bar(bar3,Others,w,color="gold",edgecolor="black",label = "Others")
bar4 = [i+w for i in bar3]
plt.bar(bar4,y,w,color="silver",edgecolor="black",label = "y")
plt.xticks(bar1+w/2,Categories)
plt.show()
```
